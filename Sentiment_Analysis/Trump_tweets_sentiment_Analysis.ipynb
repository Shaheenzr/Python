<p style="font-family:verdana;color:blue;font-size:20px">Finding out the total number of positive,negative,stop and other words from the text file(includes Trump word)</p>

```python
pos_words_count = 0
neg_words_count = 0
stop_words_count = 0
others_count = 0

for tweet in trump_finalized:
    if tweet in positive_word_list:
        
        pos_words_count = pos_words_count + 1
    
    elif tweet in negative_word_list:
        
        neg_words_count = neg_words_count + 1
    
    elif tweet in stop_word_list:
        
        stop_words_count = stop_words_count + 1
        
    else:
        
        others_count = others_count +1
        
print("Total no. of positive words present in the tweet" + " " + str(pos_words_count))
print("Total no. of negative words present in the tweet" + " " + str(neg_words_count))
print("Total no. of stop words present in the tweet" + " " + str(stop_words_count))
print("Total no. of other words present in the tweet" + " " + str(others_count))

          
```

<p style="font-family:verdana;color:blue;font-size:20px">Finding out the ratio of positive and negative words compared to the whole word count</p>

```python
#finding out the total word count

total_word_count = pos_words_count + neg_words_count + stop_words_count + others_count

print("Total number of words present in the tweets" + " " + str(total_word_count))


#finding the positive word ratio

positive_ratio = pos_words_count / total_word_count

final_positive_ratio = round(positive_ratio, 4)

print("Total number of positive words in comparison to whole word count in the tweets" + \
      " " + str(final_positive_ratio))


#finding the negative word ratio

negative_ratio = neg_words_count / total_word_count

final_negative_ratio = round(negative_ratio, 4)

print("Total number of negetive words in comparison to whole word count in the tweets" + \
      " " + str(final_negative_ratio))

```

<p style="font-family:verdana;color:blue;font-size:20px">Finding out the ratio of stop words compared to the whole word count</p>

```python

#finding the stop word ratio

stop_word_ratio = stop_words_count / total_word_count

final_stop_word_ratio = round(stop_word_ratio, 4)

print("Total number of stop words in comparison to whole word count in the tweets" + \
      " " + str(final_stop_word_ratio))

```

<p style="font-family:verdana;color:blue;font-size:20px">Finding out the ratio of other words compared to the whole word count</p>

```python

#finding the other word ratio

other_word_ratio = others_count / total_word_count

final_other_word_ratio = round(other_word_ratio, 4)

print("Total number of other words in comparison to whole word count in the tweets" +\
      " " + str(final_other_word_ratio))

```

<p style="font-family:verdana;color:blue;font-size:20px">Evaluating the ratio of positive word to negative word (sentiment analysis)</p>

<ul>
    <li>If the ratio is < 1, that means there are more negative words than positive words. 
Depends on how much the ratio is smaller than 1.<br>
     For example, wed say ~0.9 is weakly negative and ~0.2 is strongly negative.</li>
    <li>If the ratio is roughly = 1, then that means overall its neutral.</li>
    <li>If the ratio is > 1, that means there are more positive words than negative words. 
Depends on how much the ratio is bigger than 1.<br>
     For example, wed say ~1.1 is weakly positive and ~1.6 is strongly positive.</li>
</ul>

```python

analysis_ratio = (pos_words_count/ neg_words_count)

final_analysis_ratio = round(analysis_ratio,2)

# I am considering ratio < = 0.30 as strongly negative and ratio >= 0.70 as weakly negative
# I am considering ratio <=1.30  as weakly positive and ratio >=1.70 as strongly positive
# I am considering roughly ~ 1 ratio as neutral

print("The sentiment analysis ratio is " + str(final_analysis_ratio))


if ( final_analysis_ratio >= 0.95 and final_analysis_ratio <= 1.05 ):
    
    
    print("which means general sentiment of the tweets is neutral")
        
elif ( final_analysis_ratio < 0.95 ):
    
    print("which means there are more negative words than positive words in the tweet sample")
    
    if( final_analysis_ratio < 0.70 and final_analysis_ratio > 0.30 ):  
        
        print("Although the general sentiment of the tweets is negative, However there exists no extremities")
        
    elif( final_analysis_ratio <= 0.30 ): 
        
        print("So,the general sentiment of the tweets is strongly negative")
    
    else:
        
        print("So,the general sentiment of the tweets is weakly negative")
        
        
elif ( final_analysis_ratio > 1.05):
    
    print("which means there are more positive words than negative words in the tweet sample")
        
    if( final_analysis_ratio < 1.70 and final_analysis_ratio > 1.30):
        
        print("Although the general sentiment of the tweets is positive, However there exists no extremities")
    
    elif( final_analysis_ratio <= 1.30):
        
        print("So,the general sentiment of the tweets is weakly positive")
        
    else:
        
        print("So,the general sentiment of the tweets is strongly positive")

```
